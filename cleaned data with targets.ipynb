```python
import pandas as pd
```

```python
df= pd.read_csv("processed_data.csv", delimiter='\t')
```

```python
df = df[['AD_ID','AD_CLICKS','AD_IMPRESSIONS','AD_SPEND','AD_TARGETING_AGE','AD_TARGETING_INTERESTS','AD_TEXT', 'AD_POTENTIAL_TARGET','positivity_index','negativity_index', 'neutrality_index','compound_measure', 'flag', 'CTR']]
```

```python
# Deleting nan values from AD_TARGETTING_INTEREST column
df = df.dropna(subset=['AD_TARGETING_INTERESTS'])
```

```python
df
```

```python
# Collecting topics in dummy variables

temp_1 = df[df['AD_TARGETING_INTERESTS'].str.contains('LGBT|Homosexuality')]
temp_1 = temp_1.assign(LGBT = 1)

temp_2 = df[df['AD_TARGETING_INTERESTS'].str.contains('Muslim|Islam')]
temp_2 = temp_2.assign(Islam = 1)

temp_3 = df[df['AD_TARGETING_INTERESTS'].str.contains('Black|African')]
temp_3 = temp_3.assign(Black= 1) 

temp_4 = df[df['AD_TARGETING_INTERESTS'].str.contains('Amendement|Crime|Police|Gun|Cop|')]
temp_4 = temp_4.assign(Brutality = 1) 

temp_5 = df[df['AD_TARGETING_INTERESTS'].str.contains('Mexico|Mexican')]
temp_5 = temp_5.assign(Hispanic=1) 

temp_6 = df[df['AD_TARGETING_INTERESTS'].str.contains('Conservative|Conservatism')]
temp_6 = temp_6.assign(Conservative = 1) 
```

```python
frames = [temp_1, temp_2, temp_3, temp_4,temp_5, temp_6]
temp = pd.concat(frames)
```

```python
temp
```

```python
# Create new excel file     
export_csv = temp.to_csv ('cleaned data with target groups.csv', index = None, header=True) 
```
