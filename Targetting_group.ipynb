```python
import pandas as pd

# Reading excel file
excel_file = pd.read_excel(r'D:\Dersler\2nd semester\CSS\Cleaning_data.xlsx')


# Usuable colums
df = pd.DataFrame(excel_file, columns = ['AD_ID', 'AD_TARGETING_INTERESTS', 'AD_CLICKS','AD_IMPRESSIONS', 'AD_TEXT'])

# Deleting nan values from AD_TARGETTING_INTEREST column
df = df.dropna()

# Usuable rows for targetting groups

temp_1 = df[df['AD_TARGETING_INTERESTS'].str.contains('LGBT')]
temp_1 = temp_1.assign(TARGETTING_GROUP = 'LGBT')

temp_2 = df[df['AD_TARGETING_INTERESTS'].str.contains('Muslim')]
temp_2 = temp_2.assign(TARGETTING_GROUP = 'Muslim')

temp_3 = df[df['AD_TARGETING_INTERESTS'].str.contains('Black')]
temp_3 = temp_3.assign(TARGETTING_GROUP = 'Black') 

frames = [temp_1, temp_2, temp_3]
temp = pd.concat(frames)

# Create new excel file     
export_csv = temp.to_csv (r'D:\Dersler\2nd semester\CSS\Targetting_groups.csv', index = None, header=True) 

temp
```
